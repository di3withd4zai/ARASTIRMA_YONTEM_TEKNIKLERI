```python
# ==========================================================
# MP3 DATA ANALYSIS - Müzikte Cinsiyetçi Söylemler Analizi
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

# ----------------------------------------------------------
# 1) VERİYİ YÜKLEME
# ----------------------------------------------------------

df = pd.read_csv("data_mp3.csv")
df.head()

# ----------------------------------------------------------
# 2) SÜTUN İSİMLERİNİ TEMİZLEME
# ----------------------------------------------------------

rename_map = {
    'Zaman damgası': 'timestamp',
    '  Yaşınız   ': 'yas',
    '  Cinsiyetiniz ': 'cinsiyet',
    '  En çok dinlediğiniz müzik türü  ': 'muzik_turu',
    'Müzikte cinsiyetçi söylemlerin toplum üzerinde etkili olduğunu düşünüyorum.  ': 'cinsiyetci_etki',
    'Şarkılarda kadınların aşağılanmasını rahatsız edici buluyorum.  ': 'asagilama_rahatsiz',
    'Şarkı sözleri insanların kadınlara bakışını etkileyebilir.  ': 'soz_bakis_etki',
    'Ünlülerin söylemleri/davranışları hayranları etkileyebilir.  ': 'unlu_etki',
    'Müziğin kadın düşmanlığını normalleştirdiğini düşünüyorum.  ': 'muzik_normal',
    'Günlük hayatta kadın düşmanlığına sık rastladığımı düşünüyorum.': 'gunluk_kadin_dusmanligi',
    'Cinsiyetçi müzik dinleyen kişilerin bu söylemleri ilişkilerine yansıttığını gözlemledim.  ': 'iliskiye_yansima',
    'Sizce müzik platformları (Spotify, YouTube vb.) bu tarz içerikleri filtrelemeli midir?  ': 'platform_filtre',
    ' Bugüne kadar dinlediğiniz şarkılarda kadın düşmanlığı fark ettiniz mi?  ': 'farkettin_mi',
    'Şarkılardaki cinsiyetçi söylemlere karşı daha fazla farkındalık oluşturulması gerektiğini düşünüyorum.  ': 'farkindalik_gerek',
    '  Bu konuda bir proje/çalışma yapılmasının önemli olduğunu düşünüyorum.  ': 'proje_onemli',
    'Cinsiyetçi sözler içeren şarkıları dinlemeyi bırakırım.': 'birakirim',
    'Şiddet içerikli sözler normalleştirici etki yaratıyor.  ': 'siddet_normal'
}

df = df.rename(columns=rename_map)
df.head()

# ----------------------------------------------------------
# 3) TEMEL İSTATİSTİKLER
# ----------------------------------------------------------

print("Toplam Katılımcı:", len(df))
print("\nYaş Dağılımı:\n", df['yas'].value_counts())
print("\nCinsiyet Dağılımı:\n", df['cinsiyet'].value_counts())
print("\nEn Popüler Müzik Türleri:\n", df['muzik_turu'].value_counts())

# ----------------------------------------------------------
# 4) GRAFİKLER
# ----------------------------------------------------------

plt.figure(figsize=(10,5))
df['yas'].value_counts().plot(kind='bar')
plt.title("Yaş Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Sayı")
plt.show()

plt.figure(figsize=(10,5))
df['cinsiyet'].value_counts().plot(kind='bar')
plt.title("Cinsiyet Dağılımı")
plt.xlabel("Cinsiyet")
plt.ylabel("Sayı")
plt.show()

plt.figure(figsize=(12,5))
df['muzik_turu'].value_counts().plot(kind='bar')
plt.title("En Çok Dinlenen Müzik Türü")
plt.xlabel("Tür")
plt.ylabel("Sayı")
plt.show()

# ----------------------------------------------------------
# 5) LİKERT TİPİ SORU ANALİZLERİ
# ----------------------------------------------------------

likert_cols = [
    'cinsiyetci_etki',
    'asagilama_rahatsiz',
    'soz_bakis_etki',
    'muzik_normal',
    'birakirim',
    'siddet_normal'
]

for col in likert_cols:
    print("\n---", col, "---")
    print((df[col].value_counts(normalize=True) * 100).round(1))

# ----------------------------------------------------------
# 6) KORELASYON ISI HARİTASI
# ----------------------------------------------------------

df_numeric = df.copy()

mapping = {
    "Kesinlikle katılıyorum.": 5,
    "Katılıyorum.": 4,
    "Kararsızım.": 3,
    "Katılmıyorum.": 2,
    "Kesinlikle katılmıyorum.": 1
}

for col in likert_cols:
    df_numeric[col] = df_numeric[col].map(mapping)

plt.figure(figsize=(10,8))
sns.heatmap(df_numeric[likert_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Cinsiyetçi Söylem Algısı Korelasyon Haritası")
plt.show()

# ----------------------------------------------------------
# 7) OTOMATİK SONUÇ ÖZETİ
# ----------------------------------------------------------

print("""
=====================
   SONUÇ ÖZETİ
=====================

• Katılımcıların çoğu 19–24 yaş aralığındadır.
• Kadın katılımcı sayısı biraz daha fazladır.
• Pop, Rap ve Rock/Metal en çok tercih edilen türlerdir.
• Cinsiyetçi söylemlerin toplum üzerinde etkili olduğu düşüncesi yüksek orandadır.
• Şarkılarda kadınların aşağılanması ciddi seviyede rahatsızlık yaratmaktadır.
• Katılımcıların büyük kısmı bu tarz içeriklerin normalleştirici olduğunu düşünmektedir.
• Korelasyon analizi, tutum ifadelerinin birbirini desteklediğini göstermektedir.

Bu analiz MP3 raporunun tamamı için kullanılabilir.
""")
```
