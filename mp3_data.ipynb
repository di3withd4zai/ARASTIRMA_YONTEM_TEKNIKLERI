```python
import pandas as pd
df = pd.read_csv("data_mp3.csv", sep=';')
df.columns = df.columns.str.strip().str.lower()
df['likes'] = pd.to_numeric(df['likes'], errors='coerce')
df['comments'] = pd.to_numeric(df['comments'], errors='coerce')

df.head()

print(df.groupby("platform")["likes"].mean())
print(df.groupby("platform")["comments"].mean())
```
